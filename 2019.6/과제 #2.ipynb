### 제출방법 : e-mail 또는 상시개방 서버에 올려둔 뒤 연락 (가능한 메일로 보내세요.)

이번 과제는 JS dataset을 이용합니다.

## 1. 자료 병합을 통한 성공률 향상

JS 데이터셋은 크게 3종류의 자료로 나뉘어져 있었습니다.

1. PE 헤더 및 파일에서 얻은 정보(총 68가지 항목)
2. 4-gram 상위 100개의 등장 횟수 정보
3. binary 파일을 흑백 이미지화 한 자료

실습에서는 이 자료 중 한 종류씩만 사용하는 모델을 만들었습니다.

### 1-1. 1번 자료와 2번 자료를 통합하여, 하나의 자료가 총 168가지의 항목을 가지도록 하십시오.
### 1-2. 이 통합된 자료를 이용하여 다음 모델을 활용, 악성 코드 탐지 성공률을 높여 보십시오.
#### A. Decision Tree
#### B. Random Forest
#### C. 신경망 모델(DNN)

## 2. 모델 병합(stacked model)을 통한 성공률 향상

위 1, 2, 3번 자료는 각각 신경망을 이용한 악성코드 탐지 모델을 학습시키는데 사용될 수 있습니다.

### 2-1. 각 자료로 만들어진 모델의 결과를 하나로 통합하여, 최종 탐지 결과를 내놓는 통합된 모델을 만들어 보십시오.
* 여기서는 3번 자료 때문에 신경망 모델에 한합니다.
* 결과를 통합하는 방법에는 제한이 없지만, 그 방법을 설명할 수는 있어야 합니다.
    * (예 : 3가지 결과이니 다수결을 사용했다 등)
* 모델을 너무 키워 overfitting이 발생하지 않도록 최대한 주의하시기 바랍니다.

### 2-2. 위에서 만든 통합된 모델의 결과가 각 개별 모델의 결과보다 좋은지 나쁜지 확인하고 그 이유를 설명하십시오.

## 3. 각 자료에서 중요 feature를 뽑은 뒤 2번 문제를 반복

### 3-1. 3종류의 자료 중 첫 번째와 두 번째 자료에 대해, 중요도가 높다고 판단되는 항목을 다음처럼 추려내십시오.
* 1번 자료(PE 헤더 및 파일에서 얻은 정보, 총 68가지 항목)
    * 총 10가지 항목
* 2번 자료(4-gram 상위 100개)
    * 총 20가지 항목
    
꼭 correlation을 통해 판단할 필요는 없지만, 적절한 근거를 제시해야 합니다.

### 3-2. 세 번째 자료의 크기를 적당히 축소하십시오.
* 현재 thumbnail의 해상도는 256*256입니다. 이것을 적당히 줄여보십시오.
    * Image.Image.resize 함수를 사용합니다.
* 무작정 줄이지 말고, CNN 등을 통해 탐지율이 어느 정도 유지되는 선까지만 줄여야 합니다.

### 3-3.자료의 크기를 줄인 세 자료를 이용하여, 새로운 악성 코드 탐지 모델을 만들고 평가해 보십시오.
* 세 자료를 하나로 통합해도 되고, 2번 문제처럼 각 자료를 이용하는 모델을 통합해도 됩니다.
* 모델의 overfitting이 2번 문제의 경우보다 줄어드는지, 정확도는 어느 정도 유지되는지 확인하십시오.
