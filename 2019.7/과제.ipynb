## 과제 1. Discriminator를 이용한 CPRNG-GAN 모델 작성
### 1-1. Predictor 대신에 discriminator를 사용하는 GAN 모델을 만들고 학습시키시오.

### 1-2. Predictor를 사용한 GAN(이하 Pred-GAN)과 Discriminator를 사용한 GAN(이하 Disc-GAN)를 사용하는 모델의 파라메터(모델 크기, 학습률 등)을 튜닝하고, 성능을 비교하시오.

## 과제 2. Generator를 좀 더 PRNG 답게 바꾸기

Generator를 정말 PRNG처럼 사용하려면 다음 조건을 만족시켜야 합니다.
- state를 유지하고, 기존 생성 후 변경된 state의 뒤를 이어 binary sequence를 생성할 수 있어야 함
- (re)seeding, 즉 state를 원할 때 새로운 값으로 변경할 수 있어야 함

### 2-1. 이 조건들을 만족시킬 수 있도록 generator의 구조를 자유롭게 변경하여 설계하고 학습시키시오.
- 학습시킬 때 Pred-GAN과 Disc-GAN 중 원하는 것을 사용하면 됨(둘 다 할 필요는 없음)
- 위 조건들과 동치인(equivalent) 다른 조건이 있다면, 동치임을 설명하고 그 조건을 사용해도 무방함

### 2-2. 충분히 학습된 generator를 이용하여 다음 작업을 10번 수행하시오.

* 연속된(즉 state가 이어지는) binary sequence를 100 덩이 생성 (신경망을 한 번 predict하여 생성되는 것을 1덩이로 취급)
* '연속된 binary sequence를 10덩이 생성하고 reseeding하는 작업'을 10번 반복하여 총 100덩이 생성
* 위 두 binary sequence 100덩이씩을 이어붙여 생성된 두 개의 긴 binary sequence의 pseudorandomness를 비교
    * 비교 방법은 자유
